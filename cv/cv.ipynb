```python
import pandas as pd
```

```python
dev = pd.read_parquet("./dev_data_sort.parquet")
dev.head()
```

```python
submission = pd.read_csv("./submission.csv")
submission.head()
```

```python
merged = pd.merge(dev, submission, on="id", suffixes=("_df1", "_df2"))

consistent = (merged["winner_df1"] == merged["winner_df2"]).sum()

consistency_rate = consistent / len(merged)

print(f"CV: {consistency_rate: .2%}")
```
